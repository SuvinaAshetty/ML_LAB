```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('housing_area_price.csv')
df

# Commented out IPython magic to ensure Python compatibility.
# %matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')

new_df = df.drop('price',axis='columns')
new_df

price = df.price
price

# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)

#(1) Predict price of a home with area = 3300 sqr ft

reg.predict([[3300]])

reg.coef_

reg.intercept_

#Y = m * X + b (m is coefficient and b is intercept)

3300*135.78767123 + 180616.43835616432

#(1) Predict price of a home with area = 5000 sqr ft

reg.predict([[5000]])
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Predict Canada's per capita income in 2020
df_canada = pd.read_csv('canada_per_capita_income.csv')

df_canada.rename(columns={'per capita income (US$)': 'income'}, inplace=True)
X_canada = df_canada[['year']]
y_canada = df_canada['income']

reg_canada = LinearRegression()
reg_canada.fit(X_canada, y_canada)

predicted_income_2020 = reg_canada.predict([[2020]])
print(f"Predicted per capita income in Canada for 2020: {predicted_income_2020[0]:.2f} USD")

# Plot the data and regression line
plt.scatter(X_canada, y_canada, color='blue', marker='o', label='Actual Data')
plt.plot(X_canada, reg_canada.predict(X_canada), color='red', label='Regression Line')
plt.scatter(2020, predicted_income_2020, color='green', marker='x', s=100, label='Prediction (2020)')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (USD)')
plt.title("Canada's Per Capita Income Prediction")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Predict Salary of an employee with missing values handled
df_salary = pd.read_csv('salary.csv')

df_salary.rename(columns={'YearsExperience': 'experience', 'Salary': 'salary'}, inplace=True)

# Handle missing values by filling them with median values
df_salary['experience'].fillna(df_salary['experience'].median(), inplace=True)
df_salary['salary'].fillna(df_salary['salary'].median(), inplace=True)

X_salary = df_salary[['experience']]
y_salary = df_salary['salary']

reg_salary = LinearRegression()
reg_salary.fit(X_salary, y_salary)

# Predict salary for an employee with 12 years of experience
predicted_salary_12yrs = reg_salary.predict([[12]])
print(f"Predicted salary for an employee with 12 years of experience: {predicted_salary_12yrs[0]:.2f} USD")

# Plot the salary data and regression line
plt.scatter(X_salary, y_salary, color='blue', marker='o', label='Actual Data')
plt.plot(X_salary, reg_salary.predict(X_salary), color='red', label='Regression Line')
plt.scatter(12, predicted_salary_12yrs, color='green', marker='x', s=100, label='Prediction (12 years)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (USD)')
plt.title("Employee Salary Prediction")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Function to convert experience column to numeric
def convert_to_numeric(value):
    word_to_num = {
        'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5,
        'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10,
        'eleven': 11, 'twelve': 12, 'thirteen': 13, 'fourteen': 14,
        'fifteen': 15
    }
    return word_to_num.get(value.lower(), value) if isinstance(value, str) else value

# Load hiring data
df_hiring = pd.read_csv('hiring.csv')

# Convert experience column to numeric
df_hiring['experience'] = df_hiring['experience'].apply(convert_to_numeric)

# Handle missing values by filling experience with 0 and test score with median
df_hiring['experience'].fillna(0, inplace=True)
df_hiring['test_score(out of 10)'].fillna(df_hiring['test_score(out of 10)'].median(), inplace=True)
df_hiring['interview_score(out of 10)'].fillna(df_hiring['interview_score(out of 10)'].median(), inplace=True)

# Define features and target variable
X_hiring = df_hiring[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y_hiring = df_hiring['salary($)']

# Train the model
reg_hiring = LinearRegression()
reg_hiring.fit(X_hiring, y_hiring)

# Predict salaries for given candidates
candidates = np.array([[2, 9, 6], [12, 10, 10]])
predicted_salaries = reg_hiring.predict(candidates)

for i, candidate in enumerate(candidates):
    print(f"Predicted salary for candidate with {candidate[0]} yrs experience, {candidate[1]} test score, {candidate[2]} interview score: {predicted_salaries[i]:.2f} USD")

# Plot actual vs predicted salaries
plt.scatter(y_hiring, reg_hiring.predict(X_hiring), color='blue', label='Predicted vs Actual')
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary")
plt.legend()
plt.show()
```
